```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
#We have made a dictionary so we can have all the relevant values.
```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = delivery_df.append(loop_df)
    
#We have made a delivery by delivery breakdown of the entire dataset on the basis of the dictionary created above.
```

```python
delivery_df
#checking out delivery_df data set as we did a delivery by delivery breakdown.
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
        
#Here we are trying to find the bowling team. In column 'teams' we have [batting team, bowling team]
#So, if batting team is not equal to batting team, give me bowling team.
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl, axis =1)
#we add a new column bowling_team which gives us the bowling team by applying the results of previous step. 
```

### For current score column

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
#we add new column called current_score, group matches by match_id and applu cumulative sum on runs column.
```

```python
df.current_score
#we can see our code above works fine.
```

### For Balls left columns

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
#Firstly, we make two new columns, where we split 'ball' column data. The number before decimal goes into over column.
#The number after decimal goes into ball_no column.
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
#here we add a new column called 'balls_bowled' which shows how many balls in the innings have been bowled.
#we multiply the value in over column by 6 and add the ball_no. value to the product.
#For example - 10.2 = (10*6)+2 = 62nd ball of the innings.
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
#Total number of balls in an innings is 120, we subtract the bowls bowled from 120 to get the number of balls left.
#However, there can be extras as well, in case the nummber becomes negative, like 121 balls would be 120-121=-1,
#We assign the value of zero to that.
```

### For Wickets_left columns

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
#we see that only balls on which player has been dismissed have the name, else it shows zero.
#We assign 1 to the rows where a string value is present in the player_dismissed column.
#We now add a condition, where we add the number of players_dismissed in each innings.
#add a new column called 'wickets_left' where we subtract, cumulative no. of players dismissed in distinct innings from 10.
```

```python
df
#checking our dataset.
```

### For Current Run Rate column

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
#We calculate current run rate, by multiplying current score by 6 and then divide by number of balls.
```

```python
df
#checking current runrate column.
```

### For Last five column

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
    
#we have grouped matches by unique match_ids.
#for disttinct ids, we have taken a window of 30 balls to count the last five. 
```

```python
df['last_five'] = last_five
#add a new column called 'last_five' where we add the values we found above.
```

### For Runs_x column (Total runs in an innings)

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
#
```

### Final adjustments

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
#we take the required columns into a new data set called final_df.
```

```python
final_df.dropna(inplace=True)
#we have dropped the null values of last 5 here as we cannot calculate runs for 5 overs, before 5 overs have been played.
#also, we cannot accurately predict a teams score only the basis of a small sample that is of like 2 overs or 3 overs.
```

```python
final_df.isnull().sum()
#we see all the null values have been dropped.
```

```python
final_df = final_df.sample(final_df.shape[0])
#we shuffle our data randomly to protect data from any bias.
```

# Model Building

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
#doing a train test split
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
#importing neccessary libraries to work on xgboost model building
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
#we see our r2 values is 98%, which is very good.
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
